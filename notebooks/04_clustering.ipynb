```python
import pandas as pd
import numpy as np
import umap
import hdbscan
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv("D:/My Projects/amazon review insight engine/data/Reviews.csv")
embeddings = np.load("D:/My Projects/amazon review insight engine/notebooks/embeddings.npy")
```

```python
reducer = umap.UMAP(n_neighbors=15, n_components=5)
reduced_embeddings = reducer.fit_transform(embeddings)

reduced_embeddings.shape
```

```python
clusterer = hdbscan.HDBSCAN(min_cluster_size=100)
clusters = clusterer.fit_predict(reduced_embeddings)

df["cluster"] = clusters
df["cluster"].value_counts().head()
```

```python
score = silhouette_score(reduced_embeddings, clusters)
print("Silhouette Score:", score)
```

```python
df.groupby("cluster")["Score"].mean().sort_values()
```

```python
cluster_id = 2
df[df["cluster"] == cluster_id]["full_text"].sample(5)
```
