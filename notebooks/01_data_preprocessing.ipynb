```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_colwidth", 200)
```

```python
df = pd.read_csv("D:/My Projects/amazon review insight engine/data/Reviews.csv")
print("Shape:", df.shape)
df.head()
```

```python
n = min(50000, len(df))
df = df.sample(n=n, random_state=42)
df = df.dropna(subset=["Text"])
df = df[df["Text"].str.len() > 40]

df.shape
```

```python
df["Time"] = pd.to_datetime(df["Time"], unit="s")
df["YearMonth"] = df["Time"].dt.to_period("M")
```

```python
df["full_text"] = df["Summary"].fillna("") + ". " + df["Text"]
df[["Score", "full_text"]].head()
```

```python
df.to_csv("D:/My Projects/amazon review insight engine/data/Reviews.csv", index=False)
```
