```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:/My Projects/amazon review insight engine/data/Reviews.csv")
df["YearMonth"] = df["YearMonth"].astype(str)
```

```python
trend = df.groupby(["YearMonth", "cluster"]).size().unstack(fill_value=0)
trend.head()
```

```python
trend.plot(figsize=(14,6))
plt.title("Complaint Trends Over Time")
plt.show()
```
